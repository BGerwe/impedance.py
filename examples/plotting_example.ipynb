```python
import matplotlib.pyplot as plt
import numpy as np

import sys
sys.path.append('../../')

from impedance.circuits import DefineCircuit
```

```python
# read data
data = np.genfromtxt('../data/exampleData.csv', delimiter=',')

frequencies = data[:,0]
Z = data[:,1] + 1j*data[:,2]

frequencies = frequencies[np.imag(Z) < 0]
Z = Z[np.imag(Z) < 0]
```

```python
circuit = DefineCircuit(initial_guess=[.01, .005, .1, .005, .1, .001, 200], circuit='R_0-p(R_1,C_1)-p(R_1,C_1)-W_1/W_2')

circuit.fit(frequencies, Z)

print(circuit)
```

```python
# plot the data and the fit
circuit.plot(frequencies, Z, CI=True)
```
